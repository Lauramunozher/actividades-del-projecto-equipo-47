Para el proyecto final, se eligió un dataset encontrado en kaggle.com que contiene información recolectada por el Centro para el Control y Prevención de Enfermedades (CDC) de EE.UU. como parte de una encuesta telefónica realizada en 2020. El dataset contiene características de distintos individuos, así como una etiqueta para indicar si tienen una enfermedad del corazón. Las características de los individuos incluyen género,edad, raza, IMC, entre otras.

Nombre del conjunto de datos: Key Indicators of Heart Disease

Organización: Centers for Disease Control and Prevention (CDC)

Enlace de internet: https://www.kaggle.com/datasets/kamilpytlak/personal-key-indicators-of-heart-disease

El dataset contiene las siguientes columnas o datos:

* HeartDisease: "Yes" si las personas han reportado tener una enfermedad coronaria o un infarto agudo de miocardio, "No" en caso contrario
* BMI: Indice de Masa Corporal (IMC)
* Smoking: "Yes" si la persona ha fumado más de 100 cigarros en su vida, "No" en caso contrario
* AlcoholDrinking: "Yes" si es hombre adulto y toma más de 14 tragos por semana o si es mujer adulto y toma más de 7 tragos por semana, "No" en caso contrario
* Stroke: "Yes" si la persona ha tenido un derrame cerebral, "No" en caso contrario
* PhysicalHealth: Número de días que la persona ha estado enferma o herida en los últimos 30 días
* MentalHealth: Número de días que la persona no ha estado bien en cuanto a su salud mental (en los últimos 30 días)
* DiffWalking: "Yes" si la persona tiene dificultad para subir escaleras o caminar, "No" en caso contrario
* Sex: "Female" para femenino o "Male" para masculino
* AgeCategory: Categoría de edad (en rangos de 5 años)
* Race: Raza o etnicidad
* Diabetic: "Yes" si la persona ha sido diagnosticada con diabetes, "No" en caso contrario
* PhysicalActivity: "Yes" si la persona reportó realizar actividad física o ejercicio durante los últimos 30 días (sin contar su trabajo regular), "No" en caso contrario
* GenHealth: Variable categórica que describe la salud general de los encuestados
* SleepTime: Promedio diario de horas de sueño
* Asthma: "Yes" si tiene asma, "No" en caso contrario
* KidneyDisease: "Yes" si el encuestado ha sido diagnosticado con insuficiencia renal (sin contar piedras en el riñón, infección de vejiga o incontinencia), "No" en caso contrario
* SkinCancer: "Yes" si el encuestado ha sido diagnosticado cáncer de piel, "No" en caso contrario
