## Parte 1: Fundamentos de bases de datos

1) ¿Qué es un Data Lake?

Es el espacio en donde tenemos toda la información sin ser procesada para que posteriormente los analistas o ingenieros de datos sean los encargados de transformarla.
Aquí podemos contener la información de todas nuestras aplicaciones, redes sociales, aplicaciones empresariales, entre otros.

2) ¿Qué es un Data Warehouse?

Es un lugar en donde se almacena toda la información necesaria para describir la operación de un proceso y que generalmente tiene diversos orígenes, como en el caso de un negocio en línea en donde la información puede venir de Google Analytics, de la pasarela de pagos o de Shopify y nosotros más adelante decidiremos de qué manera relacionaremos esta información para su análisis. 

Mantener un almacén de datos es una tarea complicada ya que la mayoría de las veces los ingenieros de datos son los encargados de crear procesos para preservar la gobernabilidad de esta información como el tratamiento de datos atípicos, la ausencia de ellos o incluso barreras para impedir tipos de datos que no hagan sentido.

3) ETL (Extract, Transform, Load) vs ELT (Extract, Load, Transform)

Para comprender esto de una manera muy sencilla podemos definir que ETL es tomar la información de nuestro sitio web, acomodarla de manera estructurada en una base de datos y cargarla en un lugar como en Snowflake para ser consumida por nuestros usuarios de acuerdo con sus necesidades, esto implica un gran uso de recursos y conforme el negocio va creciendo se va haciendo más difícil la escalabilidad y es aquí cuando entra ELT cuyo principal objetivo es que todo este procesamiento se haga a través de herramientas en la nube como Hadoop y Spark que funcionan como clústeres haciendo el procesamiento más ligero.

## Parte 2: Selección y limpieza de los Datos en Python

Importamos librerías

```python
import pandas as pd
import numpy as np
```

Leemos los datos

```python
df  = pd.read_csv('https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv', index_col=0)
df
```

Se revisa si hay datos faltantes en cualquiera de las variables. Los resultados indican que todas las variables tienen valores faltantes a excepción de la variable X1:

```python
df.isnull().any()
```

Se explora la opción de eliminar todos los renglones que tengan al menos un valor nulo. Sin embargo, no se toma como la final para poder explorar otras opciones.

```python
df_dropna= df.dropna()
df_dropna
```

Se intenta eliminar solo las observaciones donde todas las variables estén vacías, pero no cambia el dataset original

```python
df_dropna= df.copy().dropna(how='all')
df_dropna
```

Revisamos si hay alguna columna que tenga demasiados valores nulos, pero vemos que ninguno tiene más de 20.

```python
df_dropna= df.copy().dropna(thresh = 29980, axis=1)
df_dropna
```

Debido a que no son muchas las observaciones que tienen valores nulos (42 de 30,000), sólo quitamos las 3 observaciones donde la variable objetivo faltaba. Para el resto de los valores faltantes, vamos a la opción de reemplazarlos.

```python
df_na_treatment= df.copy().dropna(subset= ['Y'])
df_na_treatment
```

De acuerdo con la descripción de las variables que se encuentra en: https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/dataset_info.txt, podemos saber que las variables X2, X3, X4, X6, X7, ..., X11 son categóricas por lo que usaremos la moda para tratar sus valores faltantes.

```python
df_na_treatment.fillna({'X2':df_na_treatment.X2.mode()[0], 'X3':df_na_treatment.X3.mode()[0], 
                           'X4':df_na_treatment.X4.mode()[0], 'X6':df_na_treatment.X2.mode()[0], 
                           'X7':df_na_treatment.X4.mode()[0], 'X8':df_na_treatment.X2.mode()[0],
                           'X9':df_na_treatment.X4.mode()[0], 'X10':df_na_treatment.X2.mode()[0],
                           'X11':df_na_treatment.X4.mode()[0]}, 
                       inplace=True)
```

Revisamos las estadísticas del resto de las variables:

```python
pd.set_option('display.max_columns', None)
df_na_treatment.drop(['X2','X3','X4','X6','X7','X8','X9','X10','X11','Y'], axis=1).describe()
```

En el caso de la edad (X5) la media y mediana parecen tener valores parecidos, mostrando una distribución no afectada por valores atípicos.
Por esa razón, se usó la media para los 5 valores faltantes de esta variable.

```python
df_na_treatment.fillna({'X5':df_na_treatment.X5.mean()}, 
                       inplace=True)
```

Para el resto de las variables (X1, X6, X7, X8, ..., X23), se utilizó la mediana:

```python
df_na_treatment.drop(['X2','X3','X4','X5','X6','X7','X8','X9','X10','X11','Y'], axis=1).fillna(df_na_treatment.drop(['X2','X3','X4','X5','X6','X7','X8','X9','X10','X11','Y'], 
                                                                                    axis=1).median(), 
                       inplace=True)
```

Se le cambió el nombre a algunas variables para que fuera más fácil su identificación y análisis.

```python
df_na_treatment.rename(columns = {'X1' : 'Given_Credit', 'X2': 'Gender', 'X3': 'Education', 'X4': 'Marital_Status',
                       'X5' : 'Age', 'Y':'Default'}, 
            inplace = True)
df_na_treatment
```
