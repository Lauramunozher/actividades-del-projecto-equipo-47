```python
df.columns
```

```python
df_na_treatment= df.copy().dropna(subset= ['Y'])
df_na_treatment
```

De acuerdo con la descripción de las variables que se encuentra en: https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/dataset_info.txt, podemos saber que las variables X2, X3 y X4 son categóricas por lo que usaremos la moda para tratar sus valores faltantes.

```python
df_na_treatment.fillna({'X2':df_na_treatment.X2.mode()[0], 'X3':df_na_treatment.X3.mode()[0], 
                           'X4':df_na_treatment.X4.mode()[0]}
                       , inplace=True)
```
